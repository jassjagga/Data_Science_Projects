```python
import numpy as np 
import cv2
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread("./test_images/images.jpg")
img.shape
```

```python
plt.imshow(img)
```

```python
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
plt.imshow(gray,cmap="gray")
```

```python
gray
```

```python
face_cascade=cv2.CascadeClassifier("C:\github\Data_Science_Projects\P03_Image_Classification_Bot\image_classifivation\model\opencv\haarcascades\haarcascade_frontalface_default.xml")
eye_cascade=cv2.CascadeClassifier("C:\github\Data_Science_Projects\P03_Image_Classification_Bot\image_classifivation\model\opencv\haarcascades\haarcascade_eyes.xml")

faces=face_cascade.detectMultiScale(gray,1.3,5)
faces
```

```python
(x,y,w,h)=faces[0]
x,y,w,h
```

```python
face_img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
plt.imshow(face_img)
```

```python
cv2.destroyAllWindows()
for(x,y,w,h)in faces:
    face_img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray=gray[y:y+h,x:x+w]
    roi_color=face_img[y:y+h,x:x+w]
    eyes=eye_cascade.detectMultiScale(roi_gray)
    for(ex,ey,ew,eh)in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew+ey+eh),(0,255,0),2)

plt.figure()
plt.imshow(face_img,cmap="gray")
plt.show()
```
