```python
#face_cascade=cv2.CascadeClassifier("C:\github\Data_Science_Projects\P03_Image_Classification_Bot\image_classifivation\model\opencv\haarcascades\haarcascade_frontalface_default.xml")
#eye_cascade=cv2.CascadeClassifier("C:\github\Data_Science_Projects\P03_Image_Classification_Bot\image_classifivation\model\opencv\haarcascades\haarcascade_eyes.xml")

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
faces=face_cascade.detectMultiScale(gray,1.3,5)
faces
```

```python
(x,y,w,h)=faces[0]
x,y,w,h
```

```python
face_img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
plt.imshow(face_img)
```

```python
cv2.destroyAllWindows()
for(x,y,w,h)in faces:
        center = (x + w//2, y + h//2)
        frame = cv2.ellipse(img, center, (w//2, h//2), 0, 0, 360, (255, 0, 255), 4)
        faceROI = gray[y:y+h,x:x+w]
        #-- In each face, detect eyes
        eyes = eye_cascade.detectMultiScale(faceROI)
        for (x2,y2,w2,h2) in eyes:
            eye_center = (x + x2 + w2//2, y + y2 + h2//2)
            radius = int(round((w2 + h2)*0.25))
            frame = cv2.circle(frame, eye_center, radius, (255, 0, 0 ), 2)

                
plt.figure()
plt.imshow(frame, cmap="gray")
plt.show()
```

```python
def get_cropped_image_if_2_eyes(image_path):
    img=cv2.imread(image_path)
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,1.3,5)
    for (x,y,w,h) in faces:
        roi_gray=gray[y:y+h,x:x+w]
        roi_color=img[y:y+h,x:x+w]
        eyes=eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >=2:
            return roi_color
```

```python
orginal_image=cv2.imread("./test_images/images.jpg")
plt.imshow(orginal_image)
```

```python
croped_image=get_cropped_image_if_2_eyes("./test_images/images.jpg")
plt.imshow(croped_image)
```
